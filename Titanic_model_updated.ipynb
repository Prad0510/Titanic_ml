```python
import pandas as pd
import numpy as np
traindata = pd.read_csv('train.csv')
testdata = pd.read_csv('test.csv')
traindata.info()
```

```python
traindata.describe()
```

```python
traindata['Age'] = traindata['Age'].fillna(traindata['Age'].median())
```

```python
traindata['Embarked'] = traindata['Embarked'].fillna(traindata['Embarked'].mode()[0])
```

```python
traindata.isnull().sum()
```

```python
traindata['Sex']=traindata['Sex'].map({'male':0, 'female':1})
```

```python
traindata.head()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')

enc.fit(traindata[['Embarked']])

encoded_embarked = enc.transform(traindata[['Embarked']]).toarray()

encoded_df = pd.DataFrame(encoded_embarked, columns=enc.get_feature_names_out(['Embarked']))

traindata = pd.concat([traindata, encoded_df], axis=1)

traindata.drop('Embarked', axis=1, inplace=True)

traindata.head()
```

```python
x=traindata[['Pclass', 'Sex', 'Age', 'Embarked_C','Embarked_Q','Embarked_S']]
y=traindata['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(x, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)


model_re_trained = LogisticRegression(solver='liblinear')
model_re_trained.fit(X_train, y_train)

y_pred = model_re_trained.predict(X_val)

accuracy = accuracy_score(y_val, y_pred)
print(f"Accuracy on the validation set: {accuracy:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(random_state=42)
random_forest_model.fit(X_train, y_train)

y_pred_rf = random_forest_model.predict(X_val)

accuracy_rf = accuracy_score(y_val, y_pred_rf)
print(f"Accuracy on the validation set (Random Forest): {accuracy_rf:.4f}")
```
